```python
"""
    We are going to build a linear regression and random forest model on two  sets of data
    1) Data obtained by omitting rows with any missing values
    2) Data obtained by inputting the missing values
"""
```

```python
#1) OMITTING MISSING VALUES
cars_omit=cars.dropna(axis=0)
```

```python
#converting categorical variables to dummy variables
cars_omit=pd.get_dummies(cars_omit,drop_first=True)
print(cars_omit)
```

```python
#importing necessary libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
#MODELBUILDING WITH OMITTED DATAS
```

```python
#Separating input and output features
x1=cars_omit.drop(["price"],axis="columns",inplace=False)#input
#inplace= False is used to not affect the original cars_omit values
y1=cars_omit["price"]#output
print(x1)
```

```python
print(y1)
```

```python
#plotting the variable price
prices=pd.DataFrame({"1. Before":y1, "2. After":np.log(y1)})
print(prices)
```

```python
prices.hist()#log is taken prevously to make the visual nice
```

```python
#Transfrom prices as a ligarithmic value
y1=np.log(y1)
```

```python
#setting the data into test and train
x_train, x_test, y_train, y_test = train_test_split(x1,y1,test_size=0.3,random_state=3)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
#test_size=0.3 mean the 70% data will go to tain set and the remaing 30% data will be retained 
#as a test set
```

```python
#BASELINE MODEL FOR OMITTING DATA
"""
We are making a base model by using test data mean value
This is to set a benchmark and to compare with our regression model
"""
```

```python
##finding the mean for test data value
base_pred=np.mean(y_test)
print(base_pred)
```

```python
# Repeating the base_pred value till lenth of the test_data(y_test)
base_pred=np.repeat(base_pred,len(y_test))
print(base_pred)
```

```python
print(help(mean_squared_error))
```

```python
#finding the RMSE(Root Mean Square Error)
base_root_mean_square_error=np.sqrt(mean_squared_error(y_test,base_pred))
print(base_root_mean_square_error)
#objective is any other model that is built in future should give less than this rmse value
```

```python
#LINEAR REGRESSION WITH OMITTED DATA
```

```python
#Setting intercept as true
lgr=LinearRegression(fit_intercept=True)
print(lgr)
```

```python
#Model
model_lin1=lgr.fit(x_train,y_train)
print(model_lin1)
```

```python
#predicting model on test set
cars_predictions_lin1=lgr.predict(x_test)
print(cars_predictions_lin1)
```

```python
#computing MSE(mean square error) and RMSE(root mean square error)
lin_mse1=mean_squared_error(y_test,cars_predictions_lin1)
lin_rmse1=np.sqrt(lin_mse1)
print(lin_rmse1)# Now we observe that there is decrease in rmse value
```

```python
#R squared value(it lets u how good is your model to explain the variability in the y)
r2_lin_test1=model_lin1.score(x_test,y_test)
r2_lin_train1=model_lin1.score(x_train,y_train)
print(r2_lin_test1,"\n",r2_lin_train1)
#Here the test set and the train set scores are nearly closer and train set is some what better
```

```python
#Regression diagnostics= Residualplot analysis
residuals1=y_test-cars_predictions_lin1#difference bettween predicted and original value
sns.regplot(x=cars_predictions_lin1,y=residuals1,scatter=True,fit_reg=False,data=cars)
#This plot tells us the all the predicted values are closer to the original
```
