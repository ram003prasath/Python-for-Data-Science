```python
import os
import matplotlib.pyplot as plt
#To work with data frames
import pandas as pd

#To perform numerical operations
import numpy as np

#To visualize data
import seaborn as sns

#Importing library for logictic regression
from sklearn.linear_model import LogisticRegression
#to use train_test_split function
from sklearn.model_selection import train_test_split
#Importing performance metrics - accuracy score and confusion
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

os.chdir("D:\\program\\PYTHON\\CSV")
incomedata=pd.read_csv("income.csv")
data=incomedata.copy()
data2=data.dropna(axis=0)
```

```python
## LOGISTIC REGRESSION
# mapping <=50,000 to 0 and >50,000 to 1 
data2["SalStat"]=data2["SalStat"].map({" less than or equal to 50,000":0," greater than 50,000":1})
print(data2)
```

```python
print(data2["SalStat"].unique())
```

```python
new_data=pd.get_dummies(data2,drop_first=True)
print(new_data)
```

```python
#storing the column names
columns_list=list(new_data.columns)
print(column_list)
```

```python
#separating the input names from data
features=list(set(columns_list)-set(["SalStat"]))
print(features)
print(len(features))
```

```python
#Storing the output values in y
y=new_data["SalStat"].values
print(y)
```

```python
#Storing the values from input features
x=new_data[features].values
print(x)
```

```python
#Splitting the data into train and test
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
print(train_x,"\n\n",test_x,"\n\n",train_y,"\n\n",test_y)
```

```python
#make an instance of the model
logistic=LogisticRegression()
print(logistic)
```

```python
#fitting the values for x and y
logistic.fit(train_x,train_y)
```

```python
logistic.coef_
```

```python
logistic.intercept_
```

```python
#prediction from test data
prediction=logistic.predict(test_x)
print(prediction)
```

```python
#confusion matric- It gives you the both no.of correct and incorrect predictions
confusion_matrix=confusion_matrix(test_y,prediction)
print(confusion_matrix)
```

```python
#calculating the accuracy
score=accuracy_score(test_y,prediction)
print(score)
```

```python
# printing the misclassified values from prediction

print("Misclassified samples: %d" %(test_y != prediction).sum())
```

```python
#reindexing the salary status names to 0,1
data2["SalStat"]=data2["SalStat"].map({" less than or equal to 50,000":0," greater than 50,000":1})
print(data2["SalStat"])
```

```python
cols=["gender","nativecountry","race","JobType"]
new_data=data2.drop(cols,axis=1)
print(new_data)
```

```python
new_data=pd.get_dummies(new_data,drop_first=True)
print(new_data)
```

```python
#storing the column names
columns_list=list(new_data.columns)
print(columns_list)
```

```python
#Separating the input names from data
features=list(set(columns_list)-set(["SalStat"]))
print(features)
```

```python
#Storing the output valus to y
y=new_data["SalStat"].values
print(y)
```

```python
#Storing the values from the input features
x=new_data[features].values
print(x)
```

```python
#Splitting the data into train and test
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=False)

print(train_x,"\n\n",test_x,"\n\n",train_y,"\n\n",test_y)
```

```python
#Make an instance of the model
logistic=LogisticRegression()
#Fitting the values for x and y
logistic.fit(train_x,train_y)
```

```python
#prediction from test data
prediction=logistic.predict(test_x)
print(prediction)
```

```python
#Calculating the accuracy
accuracy_score=accuracy_score(test_y,prediction)
print(accuracy_score)
```

```python
#Printing the misclassified values from prediction
print("Misclassified samples: %d" %(test_y!=prediction).sum())
```

```python
#importing the library of knn
from sklearn.neighbors import KNeighborsClassifier
#importing library for plotting
import matplotlib.pyplot as plt

#Storing the k nearest neighbors classifier
KNN_classifier=KNeighborsClassifier(n_neighbors=5)

#fitting the values for x and y
KNN_classifier.fit(train_x,train_y)

#predicting the test values with model
prediction=KNN_classifier.predict(test_x)
print(prediction)
```

```python
#performance metric check
confusion_matrix=confusion_matrix(test_y,prediction)
print("\tPredicted values")
print("Original values\n",confusion_matrix)
```

```python
#Calculating the accuracy 
accuracy_score=accuracy_score(test_y,prediction)
print(accuracy_score)
```

```python
# printing the misclassified values from prediction

print("Misclassified samples: %d" %(test_y != prediction).sum())
```

```python
'''Effect of k value classifier'''
misclassifiedsample=[]
#calculating error for k values between 1 and 20
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_x,train_y)
    pred_i=knn.predict(test_x)
    misclassifiedsample.append((test_y!=pred_i).sum())
print(misclassifiedsample)
```
