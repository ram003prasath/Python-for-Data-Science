```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
os.chdir("D:\\program\\PYTHON\\CSV")
data=pd.read_csv("csv2.csv")
print(data)
```

```python
data.dropna(axis=0,inplace=True)
print(data)
```

```python
sb.set(style="darkgrid")
sb.regplot(x=data["Age"],y=data["Salary"])# by default fit_reg=True
```

```python
sb.regplot(x=data["Age"],y=data["Salary"],fit_reg=False)# when fit_reg=False
```

```python
sb.regplot(x=data["Age"],y=data["Salary"],fit_reg=False,marker="+")#marker is used
```

```python
#lmplot
sb.lmplot(x="Age",y="Salary",data=data,fit_reg=False,hue="Gender",legend=True,palette="Set1")
```

```python
#Histogram
sb.distplot(data["Age"])
```

```python
sb.distplot(data["Age"],kde=False)#Kernel density estimate(kde)
```

```python
sb.distplot(data["Age"],kde=False,bins=5)
```

```python
sb.histplot(data["Age"])
```

```python
#Bar plot
#countplot is used to represent bar plot
sb.countplot(x="Age",data=data)
```

```python
#Grouped bar plot
sb.countplot(x="Age",data=data,hue="Gender")
```

```python
#Box plot and whiskers plot
sb.boxplot(y="Salary",data=data) # horizontal line is called as whiskers
```

```python
sb.boxplot(x="Salary",data=data)
```

```python
sb.boxplot(x="Gender",y="Salary",data=data)
```

```python
#grouped box and whiskers plot 
sb.boxplot(x="Gender",y="Salary",data=data,hue=data["Age"]%2)
```
