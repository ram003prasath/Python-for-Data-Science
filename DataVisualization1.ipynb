```python

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
os.chdir("D:\\program\\PYTHON\\CSV")
data=pd.read_csv("csv2.csv")
print(data)
```

```python
#data.dropna(axis=0,inplace=True) deletes the row which has NaN valuse
```

```python
#Scatter Plot
plt.scatter(data["Age"],data["Salary"],c="red")
plt.title("Scatter plot")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
#HISTOGRAM
plt.hist(data["Age"],color="green",edgecolor="red",bins=5)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#BAR PLOT
plt.bar(data["Age"],data["Salary"],color="green",edgecolor="red")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show
```
